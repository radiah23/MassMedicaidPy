```python
# Household Class

primary_applicant = Individual(
    "Female", 
    "US Citizen", 
    "1990-05-05", 
    "MA",
    disability_status=False, 
    pregnancy_status=False,
    is_primary_caretaker=True
)

dependent1 = Dependent(
    gender="Female",
    citizenship="US Citizen",
    birthdate="2010-05-12",
    state="MA",
    relationship_to_the_applicant="Child",
    disability_status=False,
    pregnancy_status=False
)

dependent2= Dependent(
    gender="Male",
    citizenship="US Citizen",
    birthdate="2010-05-12",
    state="MA",
    relationship_to_the_applicant="Child",
    disability_status=False,
    pregnancy_status=False
)


total_income = 25000

household1 = Household(primary_applicant, [dependent1], total_income)

print(f"Household Income: ${household1.get_income()}")
print(f"Household Size: {household1.household_size()}")
print(f"Number of Dependents: {household1.dependent_count()}")
print(f"Number of Children: {household1.get_children()}")
print(f"Number of Adults: {household1.get_adults()}")
print(f"Number of Seniors: {household1.get_seniors()}")
print(f"Is Primary Applicant a Parent? {household1.is_parent()}")
print(f"Has Pregnant Members? {household1.has_pregnancy()}")
print(f"Has Disabled Members? {household1.has_disabled_member()}")
print(f"Has Seniors? {household1.has_seniors()}")

#Add Method
household1.add_dependent(dependent2)
print(f"Household size after adding a dependent: {household1.household_size()}")

```
